```python
def extract(filepath):
    _df = pd.read_excel(filepath)
    df = _df.iloc[:, [1,2,4,5]]
    return df
```

```python
df = extract(filepath)
df.head()
```

```python
df = pd.DataFrame()

for filepath in filepaths:
    _df = extract(filepath)
    df = pd.concat([df,_df])
```
